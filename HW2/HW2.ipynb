## fsize=1의 사례

```r
data_3 = subset(data, fsize==1 & male==1)
```

```r
nrow(data_3)
```

```r
reg_data3 = lm(nettfa~inc+age, data=data_3)
summary(reg_data3)
```

```r
test3 = (0.90535-1)/0.16802
test3
```

```r
b3 = coef(reg_data3)[3]
b3
```

```r
s3 = sqrt(vcov(reg_data3)[3,3])
s3
```

```r
t3 = (b3-1)/s3 # test
t3
```

```r
df3 = reg_data3$df.residual
df3
```

```r
pval = pt(t3, df3)
pval
```
